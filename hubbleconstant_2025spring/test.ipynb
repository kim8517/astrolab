```python
import numpy as np
import matplotlib.pyplot as plt

def linear_regression(xdat, ydat, intercept=True):
    XY, XX = 0, 0
    xbar, ybar = np.mean(xdat), np.mean(ydat)
    if intercept:
        ydat = ydat - ybar
        xdat = xdat - xbar
    for x, y in zip(xdat, ydat):
        XY += x * y
        XX += x * x
    m = XY / XX
    b = ybar - m * xbar if intercept else 0
    return m, b

vdata = np.random.rand(20)      # You should replace this with your actual data
ddata = np.random.rand(20)      # You should replace this with your actual data

H = linear_regression(vdata, ddata, intercept=False)
plt.figure(figsize=(10, 6))
plt.scatter(vdata, ddata, label='Data Points')
plt.plot(vdata, H[0] * vdata + H[1], color='red', label='Linear Fit')
plt.xlabel('V [km/s]')
plt.ylabel('D [Mpc]')
plt.title(f'{round(H[0], 3)} km/s/Mpc')
```
